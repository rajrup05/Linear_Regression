# The main types of machine learning are defined as follows:

**Supervised Learning:** An algorithm is trained on a *labeled dataset*, where every input has a corresponding correct output. The model learns to predict outcomes for new data by finding patterns and relationships in the provided examples.

**Unsupervised Learning**: An algorithm explores *unlabeled data* to discover hidden patterns, structures, or relationships within the data without any human supervision or predefined outcomes.

**Reinforcement Learning** : An agent learns to make optimal decisions through trial and error by interacting with an environment, receiving positive feedback (rewards) or negative feedback (penalties) for its actions to maximize a cumulative reward over time.

**Semi-Supervised Learning**: A hybrid approach that uses a small amount of labeled data combined with a large amount of unlabeled data to train models efficiently, often when manual labeling is too expensive.

**Self-Supervised Learning**: A technique where the model generates its own supervisory signals (labels) from the input data itself, typically used for pre-training models to understand the inherent structure of the data

```python
import pandas as pd
df=pd.read_csv('Employee_Data.csv')
df
```

### Feature vector extraction
  Here, columns (Department, WorkedHours, Certification, YearsExperience) are sliced then converted into a NumPy array and assigned to the variable x, which will serve as independent variables (features) for model training.

Now, let's train a new Linear Regression model with these reduced features and evaluate its performance.

```python
x=df.iloc[:,0:4].values
x
```

## Feature Transformation :
###Label encoding assigns a unique number to each category, while one-hot encoding creates binary columns for each category.

Screenshot 2025-11-21 115814.png

 When to Use What

Use Label Encoding when:
 The categories have a **meaningful order** (e.g., education level, size).
Youre using tree-based models (like decision trees or random forests) that can handle ordinal relationships.

Use One-Hot Encoding when:
The categories are **unordered**.
Youre using linear models or neural networks that might misinterpret numeric labels as ordinal.

```python
from sklearn.preprocessing import LabelEncoder
en=LabelEncoder()
x[:,0]=en.fit_transform(x[:,0])
x
```

```python
y=df.iloc[:,-1].values
y
```

**random_state** is a parameter used in many machine learning functions to control the randomness of operations like data splitting, shuffling, or model initialization.

Sets a seed for the random number generator.
Ensures reproducibility  the same results every time you run the code.

Using random_state=42 (or any fixed number like 0) ensures that the split is the same every time you run the code.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

### **Mean Absolute Error (MAE)**

**MAE** measures the average magnitude of the errors in a set of predictions, without considering their direction. It's the average of the absolute differences between the predicted values and the actual values.

- **Interpretation:** A lower MAE indicates a better fit of the model to the data. It gives you a direct sense of the average error in the same units as the target variable.

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,y_pred))
```

### **R-squared (R2) Score**

The R-squared score (coefficient of determination) is a key metric for regression models. It indicates the proportion of the variance in the dependent variable that can be predicted from the independent variables.

- **Value Range:** Typically between 0 and 1, but can be negative if the model is worse than simply predicting the mean of the target variable.
- **Interpretation:**
    - **1:** The model perfectly explains all the variance in the target variable.
    - **0:** The model explains none of the variance in the target variable.
    - **Higher values** generally indicate a better fit, but context is important.

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(f"R-squared score: {r2}")
```

```python
# Create a new DataFrame with only 'YearsExperience' and 'Certification'
df_new_features = df[['YearsExperience', 'Certification']].copy()

# Convert to numpy array
x_new = df_new_features.values
y_new = df['Salary'].values

# Apply Feature Scaling to x_new
sc_new = StandardScaler()
x_new_scaled = sc_new.fit_transform(x_new)

# Split the new dataset into training and testing sets
x_train_new, x_test_new, y_train_new, y_test_new = train_test_split(x_new_scaled, y_new, test_size=0.2, random_state=0)

print("New feature set (x_new_scaled) head:")
print(x_new_scaled[:])
print("Target variable (y_new) head:")
print(y_new[:])
```

```python
# Train the new Linear Regression model
lr_new = LinearRegression()
lr_new.fit(x_train_new, y_train_new)

# Make predictions on the test set
y_pred_new = lr_new.predict(x_test_new)

# Evaluate the new model
mae_new = mean_absolute_error(y_test_new, y_pred_new)
r2_new = r2_score(y_test_new, y_pred_new)

print(f"New Model MAE: {mae_new}")
print(f"New Model R-squared score: {r2_new}")

print("\n-------Previsous Model-------")
print(f"Original Model MAE: {mean_absolute_error(y_test, y_pred)}")
print(f"Original Model R-squared score: {r2_score(y_test, y_pred)}")
```
